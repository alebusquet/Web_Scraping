```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
```

```python
# Cargar las variables

years = [1930, 1934, 1938, 1042, 1946, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978, 1982, 1986, 1990,
         1994, 1998, 2002, 2006, 2010, 2014, 2018, 2022]
```

# --> FUNCIÓN PARA TRAER DATA DE LA WEB <<--

```python
# Función para extraer la información solicitada de la web

def get_matches(year):
    
    web = f'https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup'
    
    # Mandar una solicitud a la web
    response = requests.get(web)
    content = response.text
    
    # Dar formato
    soup = BeautifulSoup(content, 'html.parser')
    
    # Cargar información de los partidos "se genera una lista"
    matches = soup.find_all('div', class_='footballbox')
    
    # Guardar en listas los equipos locales, visitantes y resultados
    home = []
    score = []
    away = []

    for match in matches:
        home.append(match.find('th', class_='fhome').get_text())
        score.append(match.find('th', class_='fscore').get_text())
        away.append(match.find('th', class_='faway').get_text())
    
    # Guardar en diccionario para que este mas ordenado
    dict_futbol = {'home': home, 'score': score, 'away': away}
    
    # Mandar a df
    df_futbol = pd.DataFrame(dict_futbol)

    # Agregar columna del año del mundial
    df_futbol.insert(0, 'year', year)
    
    return df_futbol
```

-------
# -> 1) CÓDIGO PARA DATA DE UN MUNDIAL <-
-------

```python
# Ingresar el año del Mundial del cual se desean obtener todos los resultados:

año = 2022                                  # <-------------------------------------INGRESAR EL AÑO

# Llamado a la función para obtener los datos
df_futbol = get_matches(año)
df_futbol
```

-------
# -> 2) CÓDIGO PARA DATA DE TODOS LOS MUNDIALES <-
-------

```python
# Llamar a la función y crear una lista con las devoluciones
mundial_fifa = [get_matches(year) for year in years]
```

```python
# Mostrar contenido de la lista
mundial_fifa
```

```python
# Concatenar y convertir a un dataframe
df_worldcups = pd.concat(mundial_fifa, ignore_index=True)

# Crear la carpeta para guardar el archivo, si no existe.
path = "./Files"
if not os.path.exists(path):
    os.makedirs(path)

# Guardar los datos históricos en archivo
df_worldcups.to_csv('./Files/fifa_worlcups.csv', index=False)
```
